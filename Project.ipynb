## Standardizing Column Names

Here we will convert column names from camelcase to snakecase i.e convert all names to small case and add underscores between words

```python
autos.columns
```

```python
autos.rename({"yearOfRegistration":"registration_year","monthOfRegistration":"registration_month"
              ,"notRepairedDamage":"unrepaired_damage","dateCreated":"ad_created"},axis=1,inplace=True)
autos.columns
```

```python
import re
new_columns=[]
for c in autos.columns:
    #insert underscore before Capital letters
    new_columns.append(re.sub(r'(?=[A-Z])','_',c).lower())
print(new_columns)
autos.columns=new_columns
autos.rename({"power_p_s":"power_ps"},axis=1,inplace=True)
autos.columns

    
```

## Analyzing price and odometer columns

```python
new_list=[]
for p in autos["price"]:
    p=p.replace('$',"")
    p=p.replace(',',"")
    new_list.append(p)
autos["price"]=new_list
```

```python
new_list=[]
for p in autos["odometer"]:
    p=p.replace('km',"")
    p=p.replace(',',"")
    new_list.append(p)
autos["odometer"]=new_list
autos.rename({"odometer":"odometer_km"},axis=1,inplace=True)
```

```python
autos.columns
autos["odometer_km"]=autos["odometer_km"].astype(float)
autos["price"]=autos["price"].astype(float)
```

## Final Aggregation by Brand Values

We check data for only brands with over 5% of the market share

```python
autos["brand"].unique()
```

```python
filtered_brand=autos["brand"].value_counts(normalize=True)>0.05
filtered_brand=filtered_brand[filtered_brand].index
filtered_brand
        
```

Price comparison

```python
brand_mean_price={}
for b in filtered_brand:
    brand=autos[autos["brand"]==b]
    mean_price=brand['price'].mean()
    brand_mean_price[b]=int(mean_price)
print(brand_mean_price)
        
    
```

We see that BMW, Audi, Mercedes Benz are rather expensive followed by Volksvagen and Opel, Ford are the least expensive.

Mileage comparison

```python
brand_mean_mileage={}
for b in filtered_brand:
    odo=autos[autos["brand"]==b]
    mean_miles=odo["odometer_km"].mean()
    brand_mean_mileage[b]=int(mean_miles)
print(brand_mean_mileage)
```

```python
# To make a comparison in a single dataframe
bmp_series=pd.Series(brand_mean_price,name='mean_price').sort_values(ascending=False)
bmm_series=pd.Series(brand_mean_mileage,name='mean_mileage')
print(bmp_series)
print(bmm_series)
```

```python
table=pd.merge(bmp_series,bmm_series,left_index=True, right_index=True)
print(table)
```

We can see there is a relatively weak positive correlation between mean mileage and mean price. Additionally the deviation in mileage is a lot less than the deviation of price.
